## Top 5 worst affected states

```python
grouped = x1['confirmed_cases'].groupby(x1['stateFIPS']).max()
```

```python
grouped = grouped.astype('int32')
grouped.nlargest(5)
```

```python
grouped.plot(kind="bar",figsize=(12,10),title="Worst affected states")
```
